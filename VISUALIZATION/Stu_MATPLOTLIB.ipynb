### 1) Usage Guide

#### 1) Matplotlib 소개
- Matplotlib은 표를 그리는 공간인 Figure와, 도표객체인 Axes로 구성된다
- 여기에서 Axes는 주로 X,Y축으로 구성되면 3차원으로 자료를 나타내기 위해서 Z축을 추가하기도한다
- 그림을 그리는 가장 간단한 방법은 pyplot.subplots()를 활용하는 것이다

````
fig, ax = plt.subplots()
ax.plot([1,2,3,4],[1,4,2,3]); #앞의 리스트는 X축 값 뒤의 리스트는 Y축 값을 의미한다
````

- 위의 코드에서 알 수 있듯이 Matplotlib은 각 축의 값으로 리스트 값을 받는다
- 따라서 데이터프레임을 Matplotlib으로 표현하기 위해서는 각 칼럼의 값을 리스트로 변환해야 한다는 것을 의미한다

#### 2) 개념 소개
- Matplotlib을 구성하는 개념은다음과 같다
  + Figure = 그림이 그려지는 전체 공간을 의미한다. Axes를 통해 구성된다
  ``` python
    # 새로운 Figure를 만드는 방법은 아래와 같다
    fig = plt.figure() #도표가 없는 빈공간으로서 Figure
    fig, ax = plt.subplots() #하나의 도표로 구성된 Figure
    fig, axs = plt.subplots(2,2) #2행2열의 배열로 도표가 그려진 Figure
  ```
  + Axes = 데이터를 이미지로 표현한 도표이다. 아래와 같은 옵션이 제공된다
  ``` python
    set_title() #도표의 제목 설정
    set_xlabel() ; set_ylabel() #도표의 X축과 Y축의 이름 설정
    axes.Axes.set_xlim() ; axes.Axes.set_ylim() #X축과 Y축의 limit설정
  ```
  + Axis = Locator를 바탕으로 데이터를 표시하는 좌표체계이다
  + Artist = Figure에 그려지는 모든 요소들의 총칭이다

#### 3) Matplotlib 이용
- Matplotlib을 이용하는 데는 다음 2가지 방법이 있다
  + fig, ax를 먼저 구성한 뒤에 데이터 셋을 추가
  + 데이터를 바탕으로 자동으로 fig,ax를 구성

````
x = np.linspace(0,2,100) #0부터 2까지를 100개로 분할

fig, ax = plt.subplots() #하나의 도표로 구성된 figure를 먼저 구성
ax.plot(x,x, label='linear')
ax.plot(x,x**2, label = 'quadratic')
ax.plot(x,x**3, label = 'cubic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
ax.set_title('Simple Plot')
ax.legend();
````

````
# 같은 도표를 다른 방식으로 만들 수도 있다
# plt를 활용해 데이터셋을 먼저 넣어주는 방식
plt.plot(x,x, label='linear')
plt.plot(x,x**2, label = 'quadratic')
plt.plot(x,x**3, label = 'cubic')
plt.xlabel('x label') #ax와 달리 plt는 set없이 축과 타이틀을 설정할 수 있다
plt.ylabel('y label')
plt.title('Simple Plot')
plt.legend();
````

#### 2) Plot 옵션
- 다른 도표와 마찬가지로 Pyplot도 색이나 선의 종류에 옵션을 부여할 수 있다
- 이때 사용되는 용어들은 MATLAB에서 가져왔다고 한다
- Pyplot의 기본값은 'b-(Blue Sold Line)'이다
- 하지만 옵션을 지정함으로서 다음과 같이 변경가능하다

````
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') #ro = red circle을 의미
plt.axis([0, 6, 0, 20]) #plt.axis는 순차적으로 [xmin,xmax,ymin,ymax]를 의미한다
plt.show()
````

````
import numpy as np

t = np.arange(0., 5., 0.2) #0부터 5까지 0.2간격으로 분할

plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^') 
#순차적으로 x값에 t, y값에는 t, t^2, t^3을 나타냄 
plt.show()
````

- 옵션으로 특정 변수명을 활용할 수도 있다

````
data = {'a': np.arange(50), # 1부터 49까지를 a에 배정
        'c': np.random.randint(0, 50, 50), # 0부터 50까지의 수 중 랜덤으로 50개의 정수 추출
        'd': np.random.randn(50)} # 랜덤으로 50개의 실수 추출
print(data['a'])
print(data['c'])
print(data['d'])
````

````
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50) #b는 a값에 랜덤값 * 10을 더해줌
data['d'] = np.abs(data['d']) * 100 #d는 d의 절대값에 *100
print(data['b'])
print(data['d'])
````

````
plt.scatter('a', 'b', data=data) # X축은 a, Y축은 b인 산점도
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
````

````
plt.scatter('a', 'b', c='c', s='d', data=data) #색을 c리스트에 포함된 숫자값, 사이즈를 d리스트의 값으로 지정
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
````

#### 3) 범주형 변수 표현
- 범주형 변수를 Pyplot을 활용해 표현하는 것 또한 가능하다

````
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

plt.figure(figsize=(9, 3)) #전체 도표의 사이즈를 지정

plt.subplot(131) #1행 3열의 도표 중 첫번째 
plt.bar(names, values)
plt.subplot(132) #1행 3열의 도표 중 두번째 
plt.scatter(names, values)
plt.subplot(133) #1행 3열의 도표 중 세번째
plt.plot(names, values)
plt.suptitle('Categorical Plotting')
plt.show()
````

- 위의 범주형 변수 도표에서 보았듯이 Pyplot또한 하나의 Figure에 다수의 Axes를 포함할 수 있다
- 물론 이럴 때에는 plt.figure를 활용해 figure를 먼저 만들어 준 뒤 도표를 추가해야한다

````
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t) #Y값을 구하는 함수를 지정

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02) #t1과 t2에 숫자를 부여

plt.figure() #figure를 만든 뒤
plt.subplot(211) #2행1열의 도표 중 첫번째
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212) #2행1열의 도표 중 두번째
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
````

#### 4) 텍스트 추가
- Pyplot의 큰 기능 중 하나는 도표에 text를 추가할 수 있다는 것이다

````
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)
plt.hist(x, 50, density=1, facecolor='g', alpha=0.75); #plt를 활용한 히스토그램
````

- Text를 추가할 때는 다음과 같은 규칙을 따른다
- Any text element can use math text. 
- You should use raw strings (precede the quotes with an 'r'), and surround the math text with dollar signs ($), as in TeX.

````
plt.xlabel('Smarts') #x축의 이름
plt.ylabel('Probability') #y축의 이름
plt.title('Histogram of IQ') #도표의 이름
plt.text(60, .025, r'$\mu=100,\ \sigma=15$') #text의 위치와 내용
plt.axis([40, 160, 0, 0.03]) # 각축의 Limit
plt.grid(True)
plt.show()
````

### 3) Sample plots in Matplotlib
- Matplotlib으로 그릴 수 있는 기본적인 도표 유형을 확인해보자

````
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
````

#### 1) 선형그래프

````
t=np.arange(0,2,0.01) #0부터 2까지를 0.01단위로 나눔
s=1 + np.sin(2*np.pi*t)

fig, ax = plt.subplots() #하나의 도표로 구성된 Figure생성

ax.plot(t,s) #ax에 plot을 그리면 기본적으로 선 그래프를 그린다
ax.set(xlabel='time',ylabel='voltage',title='About as simple as it gets')
ax.grid()

plt.show()
````

#### 2) 다수의 도표 활용

````
#linespace = Return evenly spaced numbers over a specified interval.
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

fig, (ax1, ax2) = plt.subplots(2, 1) #2행1열의 도표를 2개 생성, ax1과 ax2라고 명명
fig.suptitle('A tale of 2 subplots') #전체의 Title

ax1.plot(x1, y1, 'o-') #x,y에 대한 선그래프를 그리되, 원마커가 있는 선
ax1.set_ylabel('Damped oscillation')

ax2.plot(x2, y2, '.-') #x,y에 대한 선그래프를 그리되, 점마커가 있는 선
ax2.set_xlabel('time')
ax2.set_ylabel('Undamped')

plt.show()
````

- 도표를 그리는 다른 방법으로 아래 코드가 있다

````
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

#figure를 생성하는 단계 없음
plt.subplot(2, 1, 1) #subplot에서 2행 1열의 도표 중 첫번째 것에 대한 설명임을 지정
plt.plot(x1, y1, 'o-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2) #subplot에서 2행 1열의 도표 중 두번째 것에 대한 설명임을 지정
plt.plot(x2, y2, '.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
````

#### 4) 막대그래프
- Bar Chart는 데이터를 한 눈에 파악하기 좋은 도표이다
- 여기에서는 Labeling된 막대그래프, 수평막대그래프에 대해 알아보도록 하자

````
labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  #라벨 수만큼 x축에 항목 생성
width = 0.35  #막대의 두께 설정

fig, ax = plt.subplots() #하나의 차트로 이루어진 Figure 생성
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')
# ax.bar을 해주면 막대그래프를 그린다
# 또한 5개에 배분된 X축의 영역의 반을 남자, 반을 여자 막대로 채웠다

ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x) 
ax.set_xticklabels(labels)
#set_xticks = set the xaxis'tick locations
#여시서 xticks는 x축에 새겨지는 글 'G1-G5'를 의미한다
ax.legend() #도표의 범례 Men,Women을 표시한다

#아래는 막대에 라벨링하는 방법
#annotate, 즉 주석을 활용하게 된다
def autolabel(rects):
    for rect in rects:
        height = rect.get_height() #막대의 높이를 height로 지정
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
````
